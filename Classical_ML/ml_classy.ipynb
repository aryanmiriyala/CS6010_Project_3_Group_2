```python
from torch_geometric.datasets import TUDataset
from torch_geometric.loader import DataLoader
import gspan_mining as gm

dataset = TUDataset(root="data", name="MUTAG")
dataset.shuffle()
```

```python
# 80/10/10 split
train_len = int(0.8 * len(dataset))
val_len = int(0.1 * len(dataset))
test_len = len(dataset) - train_len - val_len

train_dataset = dataset[:train_len]
val_dataset   = dataset[train_len:train_len + val_len]
test_dataset  = dataset[train_len + val_len:]

train_loader = DataLoader(train_dataset, batch_size=32)
val_loader   = DataLoader(val_dataset, batch_size=32)
test_loader  = DataLoader(test_dataset, batch_size=32)

num_features = dataset.num_features
num_classes  = dataset.num_classes

print("Number of Features: ", num_features, "\nNumber of Classes : ", num_classes)
```

```python
print("Size of Training   Set: ", len(train_dataset), "  Number of Classes: ",train_dataset.num_classes)
print("Size of Validation Set: ", len(val_dataset), "   Number of Classes: ",val_dataset.num_classes)
print("Size of Testing    Set: ", len(test_dataset), "   Number of Classes: ",test_dataset.num_classes)
```
